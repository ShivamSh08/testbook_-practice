### List
1) List are ordered collection of elements of same or different data types<br>
2) List elements are enclosed in square brackets [] and are separated by comma<br>
3) List is mutable (can be changed)<br>
4) Indexing and slicing is allowed in list<br>
5) List can store duplicate elements

#### Creating Empty List (list containing no elements)

```python
w1 = []      # Method-1
w2 = list()  # Method-2
print(w1)
print(w2)
print(type(w1))
print(type(w2))
```

#### Examples of List

```python
a1 = [12,13,14,23,45,56,67]             # homogeneous list - all elements are of same data type
a2 = ['Hello','Python',2.45,False,None,87]  # hetrogeneous list - elements are of different data type
print(a1)
print(len(a1))
print(a2)
print(len(a2))
type(a2)
```

#### Indexing in List

```python
r = [10,15,20,25,30,35,40,45]
print(r)
print(r[0],r[2])   # 10 20
print(r[6],r[5])   # 40 35
print(r[-1],r[-4]) # 45 30
print(r[-2],r[-7]) # 40 15
```

#### Slicing in List

```python
print(r)
print(r[3:-2])       # start=3,end=-2,step=+1
print(r[-2:-8:-2])   # start=-2,end=-8,step=-2
print(r[-4:2:-1])    # start=-4,end=2,step=-1
print(r[7:-7:-3])    # start=7,end=-7,step=-3
```

#### Iteration in list

```python
w = [12,23,34,45]
# Method 1 - Element wise Iteration
for i in w:  # i corresponds to each and every element of the list
    print(i,end=" ")
```

```python
w = [12,23,34,45]   # len(w) = 4
# Method 2 - Index wise Iteration
for i in range(len(w)): # i = 0,1,2,3
    print(f'Element at index {i} is {w[i]}')
    
# i corresponds to the index of each and every element in the list w
# i=0,w[0]=12
# i=1,w[1]=23
# i=2,w[2]=34
# i=3,w[3]=45
```

```python
w = [12,23,34,45]  # len(w) = 4
# Method 3 - Using while loop - Index wise iteration
i=0
while i <len(w):
    print(f'Element at index {i} is {w[i]}')
    i+=1
    
# i=0,0<len(w)=>True, 0,w[0]=12
# i=1,1<len(w)=>True, 1,w[1]=23
# i=2,2<len(w)=>True, 2,w[2]=34
# i=3,3<len(w)=>True, 3,w[3]=45
```

### List Properties

#### 1) List is Mutable 
List elements can be added, deleted or even updated.

```python
q = [13,15,17,19,21,24,25]
print(q)
print(q[4])
q[4] = "Python"
print(q)   # List is mutable
q[2] = 'SQL'
print(q)
```

#### 2) List can be multiplied by an integer
This process repliactes the list elements n times, where n is the int with which the list is multiplied

```python
t1 = [3,4]
print(t1*3)
t2 = ['Hello',108,'World']
print(t2*2)
```

#### 3) Two or more list can be merged (concatenated)
This involves usage of + operator

```python
g1 = [3,5,7]
g2 = [2,4,6]
g3 = [1,9]
print(g1+g2)
merge_g1g2g3 = g1+g2+g3
print(merge_g1g2g3)
```

#### 4) min(),max(),sum() functions can be applied to a list
The list should contain int or floats only

```python
# min(),max() and sum() are not list methods but they can be applied on list
h = [12,13,15,25,10,40]
print(min(h))
print(max(h))
print(sum(h))
```

#### 5) in and not in operators can be applied to list

```python
r = [10,20,30,40,50]
print(34 in r)       # False
print(40 in r)       # True
print(52 not in r)   # True
print(20 not in r)   # False
```

#### 3) insert(index,element)
Inserts an element at a specified index.<br>
Elements previously present at that index are shifted one index towards right

```python
a1 = [13,14,15,16,17,18,19]
print(a1)
a1.insert(3,'Hello')
print(a1)
a1.insert(7,'Python')
print(a1)
a1.insert(12,'Bye')
print(a1)
a1.insert(-5,50)  # excpetional case => after insertion, the index of 50 becomes -6 and not -5
print(a1)
```

#### 4) pop(index)
Removes the element at the specified index. If no index is mentioned, it removes the last element

```python
a = [23,34,45,56,67,78,90,15,10]
print(a)
a.pop(2) # Remove element present at index 2 
print(a)
a.pop(7) # Remove element present at index 7
print(a)
a.pop()  # Removes the last element
print(a)
a.pop(-2)
print(a)
a.pop(-10)  # Index Error
print(a)
```

#### 11) copy()
Creates a copy of the list. The copied object is different from the actual object.

```python
w = [5,7,9,11,13,15]
r1 = w.copy()        # shallow copy
r2 = w
print(w)
print(r1)
print(r2)
```

```python
w[2] = "Ritesh"
print(w)    # original list
print(r1)   # copied object
print(r2)   # same object
```

```python
print(id(w),id(r1))
print(id(w),id(r2))

# 1) w and r1 are pointing to different obejcts
# 2) w and r2 are pointing to the same object
# 3) id() returns the memory address
# 4) if any change is made in w or r2, then w and r2,both get impacted.
# 4) if any change is made in w, only w and r2 gets impacted, but r1 remains unaffected.
```

### List Comprehension
1) It is short hand syntax to create a list<br>
2) It is used to create one-liner code for generating a list based on specified logic.<br>

Syntax<br>
<b> a = [expression for item in iterable if condition==True]</b>

#### How to populate a list from scratch

```python
r = []
for i in range(2,7):  # i=2,3,4,5,6
    r.append(i)
print(r)
```

```python
r_new = [i for i in range(2,7)]   # i=2,3,4,5,6
print(r_new)
```

#### Case 1 - Using if with list comprehension

```python
t = [10,13,15,22,45,654,23,45,56]   # len(t)=9
r = []
for i in range(len(t)): # i=0,1,2,3,...8
    if i%2!=0:
        r.append(t[i])
print(r)
# i=0,if=>False
# i=1,if=>True,r.append(t[1]) => r=[13]
# i=2,if=>False
# i=3,if=>True,r.append(t[3]) => r=[13,22]
# i=4,if=>False
# i=5,if=>True,r.append(t[5]) => r=[13,22,654]
# i=6,if=>False
# i=7,if=>True,r.append(t[7]) => r=[13,22,654,45]
# i=8,if=>False
```

```python
t = [10,13,15,22,45,654,23,45,56]
res1 = [t[i] for i in range(len(t)) if i%2!=0]
print(res1)
```

#### Case 2 - Using 2 for loops in list comprehension

```python
r = []
for i in range(2,5):
    for j in range(5,20,5):
        r.append(i*j)
print(r)
# r=[]
# i=2,j=5,10,15 => r=[10,20,30]
# i=3,j=5,10,15 => r=[10,20,30,15,30,45]
# i=4,j=5,10,15 => r=[10,20,30,15,30,45,20,40,60]
```

```python
new_r = [i*j for i in range(2,5) for j in range(5,20,5)]
print(new_r)
```

#### Case 3 - Using if-else in List comprehension

```python
t = [10,13,15,22,45,30,23]
r = []
for i in range(len(t)):  # len(t)=7,i=0,1,2,3,4,5,6
    if i%2==0:           
        r.append(t[i]+7)
    else:
        r.append(t[i]*3)
print(r)
# i=0,if=>True,r[0]+7=10+7=17=> r=[17]
# i=1,else=>True,r[1]*3=13*3=39=> r=[17,39]
# i=2,if=>True,r[2]+7=15+7=22=> r=[17,39,22]
# i=3,else=>True,r[3]*3=22*3=66=> r=[17,39,22,66]
# i=4,if=>True,r[4]+7=45+7=52=> r=[17,39,22,66,52]
# i=5,else=>True,r[5]*3=30*3=90=> r=[17,39,22,66,52,90]
# i=6,if=>True,r[6]+7=23+7=30=> r=[17,39,22,66,52,90,30]
```

```python
t = [10,13,15,22,45,30,23]
print(len(t))
new_t = [t[i]+7 if i%2==0 else t[i]*3 for i in range(len(t))]
print(new_t)
```
