### Tuples

1) Tuple elements are represented in () and the elements are separated by comma.<br>
2) Tuple is also an ordered collection of elements of same or different data types<br>
3) Tuple is Immutable(cant be changed)<br>
4) Indexing and slicing is allowed<br>
5) Dupliacte elements are allowed<br>

#### Create Empty Tuples

```python
t1 = ()       # Method-1
t2 = tuple()  # Method-2
print(t1)
print(t2)
print(type(t1))
print(type(t2))
```

```python
w1 = (12,13,14,15,56,13)             # homogeneous tuple       
w2 = ('Hello',False,67.23,34,None)   # hetrogeneous tuple
print(w1,len(w1))
print(w2,len(w2))
```

### Tuple Indexing and Slicing
Indexing and slicing follows the same pattern as with lists or strings

```python
q1 = (10,14,12,14,15,16,17)
print(q1[0])
print(q1[4])
print(q1[-1])
print(q1[-2])
```

```python
print(q1)
print(q1[3:6])         # start=3,end=6,step=1
print(q1[-1:-6:-2])    # start=-1,end=-6,step=-2
print(q1[-1::-2])      # start=-1,step=-2
```

### Iteration in Tuple
Is is also same like that of list or string

```python
r1 = (10,20,30,40,50)
# Method-1, Element wise iteration using For loop
for i in r1:
    print(i,end=" ")
```

```python
r1 = (10,20,30,40,50)
# Method 2, Index wise iteration using For loop
for i in range(len(r1)):  # len(r1)=5,  i=0,1,2,3,4
    print(r1[i],end=" ")
# i=0,r1[0] = 10
# i=1,r1[1] = 20
# i=2,r1[2] = 30
# i=3,r1[3] = 40
# i=4,r1[4] = 50
```

```python
r1 = (10,20,30,40,50)
# Method-3, using while loop
i =0
while i <len(r1):   # len(r1)=5,i=0,1,2,3,4
    print(r1[i],end=" ")
    i += 1
# i=0,r1[0]=10,i=i+1=>i=1
# i=1,r1[1]=20,i=i+1=>i=2
# i=2,r1[2]=30,i=i+1=>i=3
# i=3,r1[3]=40,i=i+1=>i=4
# i=4,r1[4]=50,i=i+1=>i=5
# i=5=> while becomes False
```

### Tuple Properties

#### 1)  Tuple is Immutable
Tuple elements cant be modified 

```python
a1 = ('Hello',45,23.89,'Python','Java','AWS')
print(a1)
print(a1[4])
a1[4] = 'Azure'   # Tuple is immutable
print(a1)
```

#### 2) Tuple can be multiplied by an integer
This process repliactes the tuple n times, where n is the int with which the tuple is multiplied

```python
t1 = (3,4)
print(t1*3)
t2 = ('Hello',108,'World')
print(t2*2)
```

#### 3) Two or more tuples can be merged (concatenated)
This involves usage of + operator

```python
g1 = (3,5,7)
g2 = (2,4,6)
g3 = (1,9)
print(g1+g2)
merge_g1g2g3 = g1+g2+g3
print(merge_g1g2g3)
```

#### 4) min(),max(),sum() functions can be applied to a tuple

```python
h = (12,13,15,25,10,40)
print(min(h))  # min value from h
print(max(h))  # max value from h
print(sum(h))  # sum of all the elements in h
```

#### 5) in and not in operators can be applied to tuple

```python
r = (10,20,30,40,50)
print(34 in r)
print(40 in r)
print(52 not in r)
print(20 not in r)
```

### Tuple Methods

1) <b>index(element)</b> - Returns the index of the first occurence of the element from the left

2) <b>count(element)</b> - Returns the count(frequncy) of the element passed as argument

```python
a = (10,20,30,40,50,15,10,20,30,20,40,20,25)  # len(a) =13
print('Index method')
print(a.index(20))
print(a.index(50))
print(a.index(15))

print('Count method')
print(a.count(20))
print(a.count(30))
print(a.count(15))
print(a.count(44))
```
