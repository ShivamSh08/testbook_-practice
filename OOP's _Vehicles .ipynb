```python
'''1. Write a program to manage data for Honda showroom. In the showroom they want store
information about vehicles.
for each vehicle store id, name, type(LMV,HMV),price,petrolcapacity
the vehicle can be 2 wheeler or 3 wheeler or 4 wheeler
if it is 2 wheeler then store chasie number
if four wheeler then store chasie number, engine number and engine power
if 3 wheeler then store number of seats
if it 2 wheeler then it can be geared or non geared, if it geared then store number of gears
and if it is non geared then store consumption
1. add a new vehicle
2. delete the vehicle
3. display all 2 wheeler
4. display all 3 wheeler
5. display all 4 wheeler
6. display by id
7. exit'''


class NoOfWheeler():
    def __init__(self,vid=0,nm="",type="",price="",petrolcapacity="",wheelertype=""):
        self.__vid=vid
        self.__name=nm
        self.__type= type
        self.__price=price
        self.__petrolcapacity=petrolcapacity
        self.__wheelertype=wheelertype
    
    def __str__(self):
        return f"VID:{self.__vid}, Name:{self.__name}, Type:{self.__type}, Price:{self.__price} RS, Petrol Capacity:{self.__petrolcapacity} Liters, Wheeler Type:{self.__wheelertype} Wheeler"
```

```python
class twowhgeared(NoOfWheeler):
    def __init__(self,vid=0,nm="",type="",price="",petrolcapacity="",wheelertype="",chassienumber="",geared="",numberofgears=""):
        super().__init__(vid,nm,type,price,petrolcapacity,wheelertype)
        self.__chassienumber = chassienumber
        self.__geared= geared
        self.__numberofgears = numberofgears
        
    def __str__(self):
        return super().__str__()+f" Chassie number: {self.__chassienumber}, Geared : {self.__geared}, Number of gears : {self.__numberofgears}"
class twowhnon(NoOfWheeler):
    def __init__(self,vid=0,nm="",type="",price="",petrolcapacity="",wheelertype="",chassienumber="",geared="",consumption=""):
        super().__init__(vid,nm,type,price,petrolcapacity,wheelertype)
        self.__chassienumber = chassienumber
        self.__geared= geared
        self.__consumption= consumption
    
    def __str__(self):
        return super().__str__()+f" Chassie number: {self.__chassienumber}, Geared : {self.__geared}, Consumption : {self.__consumption} "
  

class threewh(NoOfWheeler):
    def __init__(self,vid=0,nm="",type="",price="",petrolcapacity="",wheelertype="",seats=""):
        super().__init__(vid,nm,type,price,petrolcapacity,wheelertype)
        self.__seats= seats

    def __str__(self):
        return super().__str__()+f" No of Seats : {self.__seats} "


class fourwh(NoOfWheeler):
    def __init__(self,vid=0,nm="",type="",price="",petrolcapacity="",wheelertype="",chassienumber="",enginenumber="",enginepower=""):
        super().__init__(vid,nm,type,price,petrolcapacity,wheelertype)
        self.__chassienumber = chassienumber
        self.__enginenumber= enginenumber
        self.__enginepower = enginepower

    def __str__(self):
        return super().__str__()+f" Chassie number: {self.__chassienumber}, Engine number : {self.__enginenumber}, Engine power : {self.__enginepower} "

```

```python

lst=[]

def add_NewVehicle():
    vid=int(input("enetr id"))
    name=input("Enetr name of the vehicle")
    type=input("Enetr type of the vehicle(LMV,HMV) in capital letters")
    price=str(input("Enter price of the vehicle"))
    petrolcapacity=str(input("Specify Petrol  Capacity of the vehicle"))
    if wheelertype == 2:
        chassienumber = int(input("Enter chassie number of the vehicle: "))
        geared=input("Is the vehicle geared y/n: ")
        if geared == "y":
            numberofgears=int(input("Enter no of gears:"))
            v=twowhgeared(vid,name,type,price,petrolcapacity,wheelertype,chassienumber,geared,numberofgears)
        else:
            consumption= int(input("Enter fuel consumption of the vehicle in liters per km: "))
            v=twowhnon(vid,name,type,price,petrolcapacity,wheelertype,chassienumber,geared,consumption)
    elif wheelertype == 3:
        seats=int(input("Enter no of seats in the vehicle: "))
        v=threewh(vid,name,type,price,petrolcapacity,wheelertype,seats)
    elif wheelertype == 4:
        enginenumber=input("Enter engine number of the vehicle: ")
        enginepower=input("Enter engine power of the vehicle: ")    
        v=fourwh(vid,name,type,price,petrolcapacity,wheelertype,enginenumber,enginepower)
    lst.append(v)
    
def display_All():
    for vall in lst:
        print(vall)

def display_by_Id(v):
    for v in lst:
        if isinstance(v, NoOfWheeler):
            print(v)

def searchById(vid):
    for vid in lst:
        if vid.get_vid()==vid:
            return vid
    else:
        return None

def deleteById(vid):
        v=searchById(vid)
        if v!=None:
            print(v)
            ch=input("Do you want to delete this vehicle y/n:")
            if ch=="y":
                lst.remove(v)
                print(f"Vehicle with ID {v} deleted successfully")
                return True
            else:
                return False
        else:
            return False



def display_All_2wh():
    for v in lst:
        if isinstance(v, twowhgeared):
            print(v)
    for v in lst:
        if isinstance(v, twowhnon):
            print(v)

def display_All_3wh():
    for v in lst:
        if isinstance(v, threewh):
            print(v)

def display_All_4wh():
    for v in lst:
        if isinstance(v, fourwh):
            print(v)    
```
